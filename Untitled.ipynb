```python
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(42)

# Number of employees
n_employees = 500

# Generate employee IDs
employee_ids = np.arange(1001, 1001 + n_employees)

# Months
months = np.random.choice(
    ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    size=n_employees
)

# Working days
total_working_days = np.random.randint(20, 26, size=n_employees)

# Days present (attendance)
days_present = np.array([
    np.random.randint(10, wd + 1) for wd in total_working_days
])

# Leave days
leave_days = total_working_days - days_present

# Attendance ratio
attendance_ratio = days_present / total_working_days

# Previous salary (â‚¹)
previous_salary = np.random.randint(20000, 80000, size=n_employees)

# Salary change (can be negative or positive)
salary_change = np.random.randint(-5000, 15000, size=n_employees)

# Current salary
current_salary = previous_salary + salary_change

# Salary change percentage
salary_change_percent = (salary_change / previous_salary) * 100

# Risk labeling logic
risk_label = []

for i in range(n_employees):
    if (
        attendance_ratio[i] < 0.6 and salary_change_percent[i] > 10
    ) or (
        leave_days[i] > 8 and salary_change_percent[i] > 8
    ):
        risk_label.append(1)
    else:
        risk_label.append(0)

# Create DataFrame
df = pd.DataFrame({
    "employee_id": employee_ids,
    "month": months,
    "total_working_days": total_working_days,
    "days_present": days_present,
    "leave_days": leave_days,
    "attendance_ratio": attendance_ratio,
    "previous_salary": previous_salary,
    "current_salary": current_salary,
    "salary_change": salary_change,
    "salary_change_percent": salary_change_percent,
    "risk_label": risk_label
})

# View dataset
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Features and target
X = df.drop(columns=["employee_id", "month", "risk_label"])
y = df["risk_label"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

lr_metrics = {
    "Model": "Logistic Regression",
    "Accuracy": accuracy_score(y_test, y_pred_lr),
    "Precision": precision_score(y_test, y_pred_lr),
    "Recall": recall_score(y_test, y_pred_lr),
    "F1-Score": f1_score(y_test, y_pred_lr)
}
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

dt_metrics = {
    "Model": "Decision Tree",
    "Accuracy": accuracy_score(y_test, y_pred_dt),
    "Precision": precision_score(y_test, y_pred_dt),
    "Recall": recall_score(y_test, y_pred_dt),
    "F1-Score": f1_score(y_test, y_pred_dt)
}
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

rf_metrics = {
    "Model": "Random Forest",
    "Accuracy": accuracy_score(y_test, y_pred_rf),
    "Precision": precision_score(y_test, y_pred_rf),
    "Recall": recall_score(y_test, y_pred_rf),
    "F1-Score": f1_score(y_test, y_pred_rf)
}
```

```python
import pandas as pd

comparison_df = pd.DataFrame([
    lr_metrics,
    dt_metrics,
    rf_metrics
])

comparison_df
```

```python
#Among the three models, Random Forest achieved the best overall performance with higher accuracy and F1-score.
#Logistic Regression performed well but struggled with non-linear relationships, while Decision Tree showed signs of overfitting. 
#Random Forest balanced bias and variance effectively,
#making it the most reliable model for payroll risk detection.
```

```python
import matplotlib.pyplot as plt

importances = rf.feature_importances_
features = X.columns

importance_df = pd.DataFrame({
    "Feature": features,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

importance_df
```

```python
plt.figure(figsize=(8,5))
plt.barh(importance_df["Feature"], importance_df["Importance"])
plt.gca().invert_yaxis()
plt.title("Feature Importance - Random Forest")
plt.xlabel("Importance Score")
plt.show()
```
